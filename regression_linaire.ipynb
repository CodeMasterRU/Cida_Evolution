```python
# from sklearn import linear_model
from scipy import stats
import numpy as np

df = pd.read_csv("deaths_all_continents_transposed.csv", sep=',')
df['id'] = df['id'].replace('1990-1999', 1995)
df['id'] = df['id'].replace('2000-2009', 2005)
df['id'] = df['id'].replace('2010-2019', 2015)

df['Monde'] = df['Monde'].str.replace(' ', '').str.replace(',', '.').astype(float)
monde = df['Monde'].head(3)
annes = df['id'].head(3)
y = monde.to_numpy()
x = annes.to_numpy().astype('float')


mymodel = np.poly1d(np.polyfit(x, y, 3))
myline = np.linspace(1994, 2016, 50)

year = 1998
predict = mymodel(year)
print(f'prediction for this year {year} is {predict}')

plt.title('World Deaths')
plt.xlabel('Year')
plt.ylabel('Deaths')
plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()


# slope, intercept, r, p, std_err = stats.linregress(x, y)

# def myfunc(x):
#   return slope * x + intercept

# mymodel = list(map(myfunc, x))

# plt.scatter(x, y)
# plt.plot(x, mymodel)
# plt.show()
```
